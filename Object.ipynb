```python
def random_s(x1,x2,l):
    #function set for random number that could change numbers and weight
    sol = random.choices([1,0],weights=[x1,x2],k = l)
    return sol

```

```python
def one_set(m):
    if m < 10:
        list = random.sample(range(1,10),m)
    else:
        list = []
        for i in range(m):
            pl = random.sample(range(1,10),1)[0]
            list.append(pl)
            
    i = 0 ; set_i = [] ; s = int(random_s(50,50,1)[0]) ; pre_profit = 0
    for i in range (m):
        ix = list[i]
        pre_profit = ix + pre_profit
        set_i.append(ix)
    # print(set_i)
    # print(pre_profit)
    
    # print(s) 
    s2 = random.sample(range(0,100),1)[0]
    if s2 == 77:
        # print('hide prize')
        profit = pre_profit + (s2 ** 2)
        # print(profit)
    else:
        if s == 0 :
            profit = pre_profit + int(s2/2)
        elif s == 1 :
            profit = pre_profit + s2 * 2
        else :
            profit = 0
            
    object = (set_i,profit)
    # print(object)
    return object

def group_set(n,m):
    i = int(n) ; l = 0 ; group = []
    while l < i:
        object = one_set(m)
        group.append(object)
        l += 1
    # print(group)
    return group
```

```python
def resource_set(data,n,m):
    Presource = []
    for i1 in range(m):
        resource = []
        for i2 in range(n):
            r1 = data[i2][0][i1]
            resource.append(r1)
        Presource.append(resource)
    # print(Presource)
    return Presource

def profits_set(data,n):
    Pprofit = []
    for i in range(n):
        p1 = data[i][1]
        Pprofit.append(p1)
    # print(Pprofit)
    return Pprofit

def Capacities_set(reso,n,m,rate):
    Pcapacities = []
    for i1 in range(m):
        total_l = sum([reso[i1][i]  for i in range(n)])
        c1 = int(total_l * rate)
        Pcapacities.append(c1)
    # print(Pcapacities)
    return Pcapacities

def data_set(data,rate):
    n1 = len(data) ; m1 = len(data[0][0])
    profit = profits_set(data,n1)
    resource = resource_set(data,n1,m1)
    cap = Capacities_set(resource,n1,m1,rate)
    # print(n1,m1)
    return profit , resource , cap
```
